# Utils

```python
# utils
def pixcelRatio(targetImage):
    total_pixels = targetImage.shape[0] * targetImage.shape[1]
    black_pixels = total_pixels - cv2.countNonZero(targetImage)
    ratio = black_pixels / total_pixels
    
    return ratio



def LCRCheck(targetImage):
    LCR = []
    
    height, width = targetImage.shape
    left_image = targetImage[:, :width//3]
    center_image = targetImage[:, width//3:2*width//3]
    right_image = targetImage[:, 2*width//3:]
    
    LCR.append(pixcelRatio(left_image))
    LCR.append(pixcelRatio(center_image))
    LCR.append(pixcelRatio(right_image))
    
    return LCR



# test result
def HTPAccTest(GT, pred):
    resultDF = GT.compare(pred)
    
    colNames = resultDF.columns
    accuracyAll = 0
    allDataNum = len(colNames)*1000 / 2 # remove other col.
    eachDataNum = resultDF.shape[1]
    
    print("Acurracy")
    for i in colNames:
        if i[1] == "other":
            continue
        eachAcc = round(sum(resultDF[i].isnull()) / len(resultDF[i]) * 100, 2)
        print(i[0]+":",eachAcc, "%", end='\t')
        
        accuracyAll += sum(resultDF[i].isnull())
    
    print()
    print("All:", round(accuracyAll/allDataNum*100, 2))
    print()
    
```

# Crop & Resize

```python
# paths
dataPath = "./data"
resizeFolderName = "/resize/"
binaryFolderName = "/binary/"
outputFolderName = "/output/"
targetExt = ".png"

# hyperparameters
targetSize = [1920, 1080] # resize

threshold = 240 #binary

#K = 12 # clustering
#attempts = 10 # clustering

# input folder
folderName = ['house', 'tree', 'person']



# crop and resize part --------------------

for HTP in folderName:
    imagesName = os.listdir(dataPath+"/"+HTP)
    count = 0
    image_id = []
    
    if not os.path.exists(dataPath+resizeFolderName+HTP):
        os.makedirs(dataPath+resizeFolderName+HTP)
    if not os.path.exists(dataPath+binaryFolderName+HTP):
        os.makedirs(dataPath+binaryFolderName+HTP)
        
    print("-------- crop and resize - ", HTP, "--------")
    tick = datetime.now()
    

    for image in imagesName:
        imagePath = dataPath+"/" + HTP + "/" + image
        
        
        if os.path.isfile(imagePath):
            if image.split('.')[1] == "jpg" or image.split('.')[1] == "png":
                image_id.append(image.split('.jpg')[0])
                newImageName = image
                
                img = cv2.imread(imagePath, cv2.IMREAD_COLOR)
                
                # crop ------------------------------------------------
                crop_img = img[731:4797, 25:6754]
                
                # resize -------------------------------------------------- + cropped
                h,w = crop_img.shape[:2]
                ash=targetSize[1]/h
                asw=targetSize[0]/w
                
                if asw<ash:
                    sizeAs=(int(w*asw),int(h*asw))
                else:
                    sizeAs=(int(w*ash),int(h*ash))
                    
                
                resizedImage = cv2.resize(crop_img,dsize=sizeAs)
                cv2.imwrite(dataPath+resizeFolderName+HTP + "/" + newImageName + "_cropped" + targetExt , resizedImage)
                
                # binary -------------------------------------------------- + cropped, resized
                gray = cv2.cvtColor(crop_img, cv2.COLOR_BGR2GRAY)
                ret, binaryImage = cv2.threshold(gray, threshold, 255, cv2.THRESH_BINARY)

                cv2.imwrite(dataPath+binaryFolderName+HTP + "/" + newImageName + "_bin" + targetExt , binaryImage)
        
        processStatus = str(count) + "/" + str(len(imagesName)-1) + "    " + str(int(count/(len(imagesName)-1)*100)) + "%"
        print("\r" + processStatus, end="")
        
        count += 1
        
        
    if HTP == "house":
        house_image_id = image_id

    elif HTP == "tree":
        tree_image_id = image_id

    elif HTP == "person":
        person_image_id = image_id
        
    tock = datetime.now()
    
    print("\t", int((tock - tick).seconds/60), "min.", (tock - tick).seconds%60, "sec.")
                
                
```

# Pred to CSV type (object detection)

```python
# object detection part --------------------
def objectDetectionChecker(HTP, count, var1, var2, var3, var4, var5):
    
    # house
    if HTP == "house":
        global house_door_yn
        global house_roof_yn
        global house_window_cnt
        if count > 0:
            if var1 > 0: # tempHouse_door_yn
                house_door_yn.append("y")
            else:
                house_door_yn.append("n")

            if var2 > 0: # tempHouse_roof_yn
                house_roof_yn.append("y")
            else:
                house_roof_yn.append("n")

            if var3 == 0: # tempHouse_window_yn
                house_window_cnt.append("absence")
            elif var3 >= 1:
                house_window_cnt.append("1 or 2")
            elif var3 >= 3:
                house_window_cnt.append("more than 3")

        
    
    # tree
    elif HTP == "tree":
        global tree_branch_yn
        global tree_root_yn
        global tree_crown_yn
        global tree_fruit_yn
        global tree_gnarl_yn
        if count > 0:
            if var1 > 0: # tempTree_branch_yn
                tree_branch_yn.append("y")
            else:
                tree_branch_yn.append("n")
                
            if var2 > 0: # tempTree_root_yn
                tree_root_yn.append("y")
            else:
                tree_root_yn.append("n")
                
            if var3 > 0: # tempTree_crown_yn
                tree_crown_yn.append("y")
            else:
                tree_crown_yn.append("n")
                
            if var4 > 0: # tempTree_fruit_yn
                tree_fruit_yn.append("y")
            else:
                tree_fruit_yn.append("n")
                
            if var5 > 0: # tempTree_gnarl_yn
                tree_gnarl_yn.append("y")
            else:
                tree_gnarl_yn.append("n")

    
    # person
    elif HTP == "person":
        global person_eye_yn
        global person_leg_yn
        global person_mouth_yn
        global person_arm_yn
        if count > 0:
            if var1 > 0: # tempPerson_eye_yn
                person_eye_yn.append("y")
            else:
                person_eye_yn.append("n")
                
            if var2 > 0: # tempPerson_leg_yn
                person_leg_yn.append("y")
            else:
                person_leg_yn.append("n")
                
            if var3 > 0: # tempPerson_mouth_yn
                person_mouth_yn.append("y")
            else:
                person_mouth_yn.append("n")
                
            if var4 > 0: # tempPerson_arm_yn
                person_arm_yn.append("y")
            else:
                person_arm_yn.append("n")
```

# Object Detection (each part)

```python
# object detection part --------------------
tempHouse_door_yn = 0
tempHouse_roof_yn = 0
tempHouse_window_cnt = 0
    
tempTree_branch_yn = 0
tempTree_root_yn = 0
tempTree_crown_yn = 0
tempTree_fruit_yn = 0
tempTree_gnarl_yn = 0
    
tempPerson_eye_yn = 0
tempPerson_leg_yn = 0
tempPerson_mouth_yn = 0
tempPerson_arm_yn = 0

ODtempSize_house = []
ODtempLoc_house = []

ODtempSize_tree = []
ODtempLoc_tree = []

ODtempSize_person = []
ODtempLoc_person = []

ODtempMinX = 1500
ODtempMaxX = 0
ODtempMinY = 1500
ODtempMaxY = 0
ODtempSize = ''
ODtempLoc = ''


def imageSquare(tempMinX, tempMaxX, tempMinY, tempMaxY, detectMinX, detectMaxX, detectMinY, detectMaxY):
    detectMinX = min(int(detectMinX), tempMinX)
    detectMaxX = max(int(detectMaxX), tempMaxX)
    detectMinY = min(int(detectMinY), tempMinY)
    detectMaxY = max(int(detectMaxY), tempMaxY)
        
    return detectMinX, detectMaxX, detectMinY, detectMaxY



for HTP in folderName:
    #imagesName = os.listdir(dataPath+resizeFolderName+HTP)
    count = 0
    startEnd = 0

    # reset values
    tempMinX = 1500
    tempMaxX = 0
    tempMinY = 1500
    tempMaxY = 0
    detectImage = ''
    
    print("-------- object detection - ", HTP, "--------")
    tick = datetime.now()
    
                
    # object detection -----------------------------------------
    if path.exists("yolov5/best_"+HTP+".pt"):
        print("Object Detection run")
        cmd = "python yolov5/detect.py --weights yolov5/best_"+HTP+".pt --conf 0.3 --imgsz 1000 --source "+ dataPath+resizeFolderName+HTP
        ODResult = subprocess.run(cmd, capture_output=True, shell=True)
        print("Object Detection end")

        for detectLine in str(ODResult).split('\\r\\n'):
            
            # detecting start and reset ---------------------------------------------------
            if detectLine == "start":
                startEnd = 1
                tempHouse_door_yn = 0
                tempHouse_roof_yn = 0
                tempHouse_window_cnt = 0
                    
                tempTree_branch_yn = 0
                tempTree_root_yn = 0
                tempTree_crown_yn = 0
                tempTree_fruit_yn = 0
                tempTree_gnarl_yn = 0
                    
                tempPerson_eye_yn = 0
                tempPerson_leg_yn = 0
                tempPerson_mouth_yn = 0
                tempPerson_arm_yn = 0

                # reset values
                tempMinX = 1500
                tempMaxX = 0
                tempMinY = 1500
                tempMaxY = 0

                continue
                
            # get size and loc., each image ---------------------------------------------------
            elif detectLine == "end":
                startEnd = 0
                count += 1
                
                tempX = 0
                tempY = 0
                if (tempMaxX - tempMinX) > 0:
                    tempX = tempMaxX - tempMinX
                else:
                    tempX = 0
                if (tempMaxY - tempMinY) > 0:
                    tempY = tempMaxY - tempMinY
                else:
                    tempY = 0
                
                targetSize = tempX * tempY
                originImage = cv2.imread(dataPath+resizeFolderName+HTP + "/" + detectImage, cv2.IMREAD_COLOR)
                SMBSize = targetSize / (originImage.shape[0] * originImage.shape[1])
                
                # create white image
                binaryImage = np.ones((originImage.shape[0], originImage.shape[1]), dtype=np.uint8) * 255
                binaryImage[tempMinX:tempMaxX, tempMinY:tempMaxY] = 0 # image area to black(zero)
                
                LCRList = LCRCheck(binaryImage)
                if sum(LCRList) > 0:
                    targetLocation = LCRList.index(max(LCRList))
                else:
                    targetLocation = '-'
                
                if HTP == "house":
                    objectDetectionChecker(HTP, count, \
                                           tempHouse_door_yn, tempHouse_roof_yn, tempHouse_window_cnt, 0, 0)
                    
                    # size
                    if SMBSize >= 0.60:
                        ODtempSize_house.append("big")
                    elif SMBSize >= 0.16:
                        ODtempSize_house.append("middle")
                    elif SMBSize > 0:
                        ODtempSize_house.append("small")
                    else:
                        ODtempSize_house.append("-")
                        
                    # loc.
                    if targetLocation == 0:
                        ODtempLoc_house.append("left")
                    elif targetLocation == 1:
                        ODtempLoc_house.append("center")
                    elif targetLocation == 2:
                        ODtempLoc_house.append("right")
                    else:
                        ODtempLoc_house.append("-")
                    
                elif HTP == "tree":
                    objectDetectionChecker(HTP, count, \
                                           tempTree_branch_yn, tempTree_root_yn, tempTree_crown_yn, tempTree_fruit_yn, tempTree_gnarl_yn)
                    
                    # size
                    if SMBSize >= 0.60:
                        ODtempSize_tree.append("big")
                    elif SMBSize >= 0.16:
                        ODtempSize_tree.append("middle")
                    elif SMBSize > 0:
                        ODtempSize_tree.append("small")
                    else:
                        ODtempSize_tree.append("-")
                        
                    # loc.
                    if targetLocation == 0:
                        ODtempLoc_tree.append("left")
                    elif targetLocation == 1:
                        ODtempLoc_tree.append("center")
                    elif targetLocation == 2:
                        ODtempLoc_tree.append("right")
                    else:
                        ODtempLoc_tree.append("-")
                    
                elif HTP == "person":
                    objectDetectionChecker(HTP, count, \
                                           tempPerson_eye_yn, tempPerson_leg_yn, tempPerson_mouth_yn, tempPerson_arm_yn, 0)
                    
                    # size
                    if SMBSize >= 0.40:
                        ODtempSize_person.append("big")
                    elif SMBSize >= 0.16:
                        ODtempSize_person.append("middle")
                    elif SMBSize > 0:
                        ODtempSize_person.append("small")
                    else:
                        ODtempSize_person.append("-")
                        
                    # loc.
                    if targetLocation == 0:
                        ODtempLoc_person.append("left")
                    elif targetLocation == 1:
                        ODtempLoc_person.append("center")
                    elif targetLocation == 2:
                        ODtempLoc_person.append("right")
                    else:
                        ODtempLoc_person.append("-")

                        
                        
                        
            # detecting from YOLOv5, each annotation -------------------------------
            if startEnd == 1:
                detectLine = detectLine.split(' ')
                detectMinX = detectLine[1]
                detectMaxX = detectLine[2]
                detectMinY = detectLine[3]
                detectMaxY = detectLine[4]
                detectImage = detectLine[5]

                
                if HTP == "house":
                    # house(5): ['roof', 'below', 'door', 'window', 'unknown']
                    if detectLine[0] == "0":
                        tempHouse_roof_yn += 1
                    elif detectLine[0] == "2":
                        tempHouse_door_yn += 1
                    elif detectLine[0] == "3":
                        tempHouse_window_cnt += 1
                        
                    # size and loc. below, roof
                    if detectLine[0] == "0": # roof
                        tempMinX, tempMaxX, tempMinY, tempMaxY = imageSquare(tempMinX, tempMaxX, tempMinY, tempMaxY, \
                                                                             detectMinX, detectMaxX, detectMinY, detectMaxY)
                    elif detectLine[0] == "1": # below
                        tempMinX, tempMaxX, tempMinY, tempMaxY = imageSquare(tempMinX, tempMaxX, tempMinY, tempMaxY, \
                                                                             detectMinX, detectMaxX, detectMinY, detectMaxY)
                            
            
                elif HTP == "tree":
                    # tree(7): ['branch','root','fruit','crown','gnarl','body','unknown']
                    if detectLine[0] == "0":
                        tempTree_branch_yn += 1
                    elif detectLine[0] == "1":
                        tempTree_root_yn += 1
                    elif detectLine[0] == "2":
                        tempTree_fruit_yn += 1
                    elif detectLine[0] == "3":
                        tempTree_crown_yn += 1
                    elif detectLine[0] == "4":
                        tempTree_gnarl_yn += 1
                        
                    # size and loc. crown, body
                    if detectLine[0] == "3": # crown
                        tempMinX, tempMaxX, tempMinY, tempMaxY = imageSquare(tempMinX, tempMaxX, tempMinY, tempMaxY, \
                                                                             detectMinX, detectMaxX, detectMinY, detectMaxY)
                    elif detectLine[0] == "5": # body
                        tempMinX, tempMaxX, tempMinY, tempMaxY = imageSquare(tempMinX, tempMaxX, tempMinY, tempMaxY, \
                                                                             detectMinX, detectMaxX, detectMinY, detectMaxY)
                    
            
                elif HTP == "person":
                    # person(8): ['eye','leg','mouth','arm','head','upper body','lower body','unknown']
                    if detectLine[0] == "0":
                        tempPerson_eye_yn += 1
                    elif detectLine[0] == "1":
                        tempPerson_leg_yn += 1
                    elif detectLine[0] == "2":
                        tempPerson_mouth_yn += 1
                    elif detectLine[0] == "3":
                        tempPerson_arm_yn += 1
                        
                    # size and loc. head, upper body, lower body
                    if detectLine[0] == "4": # head
                        tempMinX, tempMaxX, tempMinY, tempMaxY = imageSquare(tempMinX, tempMaxX, tempMinY, tempMaxY, \
                                                                             detectMinX, detectMaxX, detectMinY, detectMaxY)
                    elif detectLine[0] == "5": # upper body
                        tempMinX, tempMaxX, tempMinY, tempMaxY = imageSquare(tempMinX, tempMaxX, tempMinY, tempMaxY, \
                                                                             detectMinX, detectMaxX, detectMinY, detectMaxY)
                    elif detectLine[0] == "6": # lower body
                        tempMinX, tempMaxX, tempMinY, tempMaxY = imageSquare(tempMinX, tempMaxX, tempMinY, tempMaxY, \
                                                                             detectMinX, detectMaxX, detectMinY, detectMaxY)
                        
                        

            processStatus = str(count) + "/" + str(len(imagesName)) + "    " + str(int(count/(len(imagesName))*100)) + "%"
            print("\r" + processStatus, end="")
            
    else:
        print("There is no pre-trined model file.\t", HTP)
                
    
    tock = datetime.now()
    print("\t", int((tock - tick).seconds/60), "min.", (tock - tick).seconds%60, "sec.")
        
    
    print();print()
    
    


house_imageSMBSize_empty = [i for i, value in enumerate(ODtempSize_house) if value == '-']
house_imageLCRLoc_empty = [i for i, value in enumerate(ODtempLoc_house) if value == '-']

tree_imageSMBSize_empty = [i for i, value in enumerate(ODtempSize_tree) if value == '-']
tree_imageLCRLoc_empty = [i for i, value in enumerate(ODtempLoc_tree) if value == '-']

person_imageSMBSize_empty = [i for i, value in enumerate(ODtempSize_person) if value == '-']
person_imageLCRLoc_empty = [i for i, value in enumerate(ODtempLoc_person) if value == '-']

#print(house_imageSMBSize_empty)
#print(house_imageLCRLoc_empty)

#print(tree_imageSMBSize_empty)
#print(tree_imageLCRLoc_empty)

#print(person_imageSMBSize_empty)
#print(person_imageLCRLoc_empty)

house_imageSMBSize = ODtempSize_house
house_imageLCRLoc = ODtempLoc_house

tree_imageSMBSize = ODtempSize_tree
tree_imageLCRLoc = ODtempLoc_tree

person_imageSMBSize = ODtempSize_person
person_imageLCRLoc = ODtempLoc_person
```

# Image Processiong (size and loc.)

```python
# size and location part --------------------

for HTP in folderName:
    imagesName = os.listdir(dataPath+binaryFolderName+HTP)
    count = 0
    
    print("-------- size and loc. - ", HTP, "--------")
    tick = datetime.now()
    
    tempSize = []
    tempLoc = []
    

    for image in imagesName:
        #imagePath = dataPath+ resizeFolderName + HTP + "/" + image
        binImagePath = dataPath+ binaryFolderName + HTP + "/" + image
        
        
        if os.path.isfile(binImagePath):
            if image.split('.')[-1] == "jpg" or image.split('.')[-1] == "png":
                img = cv2.imread(binImagePath, cv2.IMREAD_GRAYSCALE)
        
        
                locSize = [100, 60]
                locSizeImage = cv2.resize(img,dsize=locSize)
                re_crop_img = locSizeImage[1:29, :] # for loc, size
                
                # calculate image size -------------------------------------
                sizeRatio = pixcelRatio(re_crop_img)
                
                if HTP == "person":
                    if sizeRatio >= 0.40:
                        tempSize.append("big")
                    elif sizeRatio >= 0.16:
                        tempSize.append("middle")
                    else:
                        tempSize.append("small")
                else:
                    if sizeRatio >= 0.60:
                        tempSize.append("big")
                    elif sizeRatio >= 0.16:
                        tempSize.append("middle")
                    else:
                        tempSize.append("small")
                
                
                # calculate image location --------------------------------
                LCRList = LCRCheck(re_crop_img)
                imageLocation = LCRList.index(max(LCRList))
                
                if imageLocation == 0:
                    tempLoc.append("left")
                elif imageLocation == 1:
                    tempLoc.append("center")
                elif imageLocation == 2:
                    tempLoc.append("right")
                    
        count += 1
        processStatus = str(count) + "/" + str(len(imagesName)) + "    " + str(int(count/(len(imagesName))*100)) + "%"
        print("\r" + processStatus, end="")
        
                    
                    
    if HTP == "house":
        house_imageLCRLoc = tempLoc
        for i in range(0,len(house_imageSMBSize_empty)):
            house_imageSMBSize[house_imageSMBSize_empty[i]] = tempSize[house_imageSMBSize_empty[i]]
            #house_imageLCRLoc[house_imageLCRLoc_empty[i]] = tempLoc[house_imageLCRLoc_empty[i]]

    elif HTP == "tree":
        tree_imageLCRLoc = tempLoc
        for i in range(0,len(tree_imageSMBSize_empty)):
            tree_imageSMBSize[tree_imageSMBSize_empty[i]] = tempSize[tree_imageSMBSize_empty[i]]
            #tree_imageLCRLoc[tree_imageLCRLoc_empty[i]] = tempLoc[tree_imageLCRLoc_empty[i]]

    elif HTP == "person":
        person_imageLCRLoc = tempLoc
        for i in range(0,len(person_imageSMBSize_empty)):
            person_imageSMBSize[person_imageSMBSize_empty[i]] = tempSize[person_imageSMBSize_empty[i]]
            #person_imageLCRLoc[person_imageLCRLoc_empty[i]] = tempLoc[person_imageLCRLoc_empty[i]]
                    
                    
    tock = datetime.now()
    print("\t", int((tock - tick).seconds/60), "min.", (tock - tick).seconds%60, "sec.")
```
