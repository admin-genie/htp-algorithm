```python
def pixcelRatio(targetImage):
    total_pixels = targetImage.shape[0] * targetImage.shape[1]
    black_pixels = total_pixels - cv2.countNonZero(re_crop_img)
    ratio = black_pixels / total_pixels
    
    return ratio



def LCRCheck(targetImage):
    LCR = []
    
    height, width, _ = targetImage.shape
    left_image = targetImage[:, :width//3]
    center_image = targetImage[:, width//3:2*width//3]
    right_image = targetImage[:, 2*width//3:]
    
    LCR.append(pixcelRatio(left_image))
    LCR.append(pixcelRatio(center_image))
    LCR.append(pixcelRatio(right_image))
    
    return LCR
```

```python
# test result
def HTPAccTest(GT, pred):
    resultDF = GT.compare(pred)
    
    colNames = resultDF.columns
    accuracyAll = 0
    allDataNum = resultDF.size
    eachDataNum = resultDF.shape[1]
    
    print("Acurracy")
    for i in colNames:
        #eachAcc = (roundresultDF[i].mean() * 100, 2)
        trueNum = sum(resultDF[i].count(True))
        eachAcc = round(trueNum/eachDataNum, 2)
        print(str(i)+":",eachAcc , end='\t')
        
        accuracyAll += trueNum
    
    print("All:", round(accuracyAll/allDataNum*100, 2))
    
```
